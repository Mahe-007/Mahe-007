```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset("car_crashes")
df.head()
```

```python
df.sort_values(["total","speeding"],ascending=[1,0])
```

```python
df["grandsum"]=df["speeding"]+df["alcohol"]
df.head()
```

```python
df=df.drop(columns=["grandsum"])
df.head()
```

```python
df.to_csv("Accidents.csv",index=False)
```
